```python
df = pd.read_excel('/content/20241220_084513_Matters_KE.xlsx', 'Matters', skiprows=3)

wb = openpyxl.load_workbook('/content/20241220_084513_Matters_KE.xlsx')
ws = wb['Matters']

colors = ['f3c4c8', 'fa818b',
          'f83e4d', 'f7a4ee']
fillers = []

for color in colors:
    temp = PatternFill(patternType='solid',
                       fgColor=color)
    fillers.append(temp)

# Date String in dd/mm/yyyy HH:MM:SS format
curr_date_str = "1/2/2025"

# Convert string to datetime object
curr_date = datetime.strptime(curr_date_str, "%m/%d/%Y")

# GETTING TO MATTERS IN CENTERBASE
driver.get("https://cohmlaw.centerbase.com/web") #INSERT URL FOR CENTERBASE - MATTERS
print(driver.title)
time.sleep(4)
dashboardLbl = driver.find_element(By.XPATH, "/html/body/div[7]/div/div/div/div/div[2]/div/div[1]/div/div[3]/div/div/div/div[1]/div[2]/h2/a")
print(dashboardLbl.text)
mattersLink = driver.find_element(By.XPATH, "/html/body/div[7]/div/div/div/div/div[1]/div/div/div/div[3]/div/a[5]")
mattersLink.click()
time.sleep(2)

for index, row in itertools.islice(df.iterrows(),340,None): # THE ROW YOU WANT TO START ON
    print(index)
    matter = row['Matter Name'][:row['Matter Name'].find(' ')]
    print(matter)
    num1 = matter[:matter.find('-')]
    num2 = matter[matter.find('-')+1:]
    if num1 != num2:
        print("Checking...")
        cell = "B{}".format(index+5)
        matterSearchBox = driver.find_element(By.XPATH, "/html/body/div[7]/div/div/div/div/div[2]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div/div/div[1]/input")
        matterSearchBox.send_keys(Keys.CONTROL + "a")
        matterSearchBox.send_keys(Keys.DELETE)
        time.sleep(1)
        matterSearchBox.send_keys(matter)
        matterSearchBox.send_keys(Keys.RETURN)
        try:
          matterBtn = WebDriverWait(driver, 30).until(
              EC.presence_of_element_located((By.XPATH, "/html/body/div[7]/div/div/div/div/div[2]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div/div/div/div[3]/div/div/div/div/div/div[3]/div/div/div/div/div/div/div/div[5]/div[2]/div/div[1]/div[3]/div"))
              )
          time.sleep(2)
          matterBtn = driver.find_element(By.XPATH, "/html/body/div[7]/div/div/div/div/div[2]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div/div/div/div[3]/div/div/div/div/div/div[3]/div/div/div/div/div/div/div/div[5]/div[2]/div/div[1]/div[3]/div")

          matterBtn.click()
          print("Found Matter")
          try:
            billingEntries = WebDriverWait(driver, 15).until(
                EC.presence_of_element_located((By.XPATH, "/html/body/div[7]/div/div/div/div/div[2]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div[2]/div/div/div/div/div/div[4]/ul/li[9]/a"))
                )
            matterLbl = driver.find_element(By.XPATH, "/html/body/div[7]/div/div/div/div/div[2]/div/div[1]/div/div[3]/div/div/div/div[1]/div[2]/h3")
            print(matterLbl.text)
            billingEntries.click()
            print("Found Billing Entries Link")
            try:
              time.sleep(2)
              dataTable = WebDriverWait(driver, 60).until(
                  EC.presence_of_element_located((By.XPATH, "/html/body/div[7]/div/div/div/div/div[2]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div[2]/div/div/div/div/div/div[4]/div[8]/div/div/div/div/div/div[3]/div/div/div/div/div/div/div/div[5]/div[2]/div/div/div[1]"))
                  )
              if (dataTable.text == "No data to display"):
                  print("No data to display")
                  raise NoSuchElementException
              print("There is data")
              latestBillingEntry = WebDriverWait(driver, 10).until(
                  EC.presence_of_element_located((By.XPATH, "/html/body/div[7]/div/div/div/div/div[2]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div[2]/div/div/div/div/div/div[4]/div[8]/div/div/div/div/div/div[3]/div/div/div/div/div/div/div/div[5]/div[2]/div/div[1]/div[3]/div"))
                  )
              time.sleep(2)
              latestBillingEntry = driver.find_element(By.XPATH, "/html/body/div[7]/div/div/div/div/div[2]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div[2]/div/div/div/div/div/div[4]/div[8]/div/div/div/div/div/div[3]/div/div/div/div/div/div/div/div[5]/div[2]/div/div[1]/div[3]/div")
              date = latestBillingEntry.get_attribute("innerHTML")
              print("Latest Billing Entry: " + date)
            except:
              date = "None"
              print("Couldn't find billing entries")
          except:
            date = "None"
            print("Couldn't find billing entries link")
        except:
          date = "None"
          print("Couldn't find matter / no results for matter")


        # PUT DATA INTO EXCEL
        ws[cell] = date
        if (date != "None"):
            dt_obj = datetime.strptime(date, "%m/%d/%Y")
            delta = relativedelta.relativedelta(curr_date, dt_obj)
            years = delta.years
            if (years >= 5):
                ws[cell].fill = fillers[2]
            elif (years >= 3):
                ws[cell].fill = fillers[1]
            elif (years >= 1):
                ws[cell].fill = fillers[0]
        elif (date == "None"):
            ws[cell].fill = fillers[3]
        wb.save("/content/20241220_084513_Matters_KE.xlsx")
        print("SUCCESSFULLY SAVED LAST BILLING ENTRY")

        # HIT THE GO BACK BUTTON TO RETURN TO THE MATTERS SEARCH BOX
        try:
          matterLbl = driver.find_element(By.XPATH, "/html/body/div[7]/div/div/div/div/div[2]/div/div[1]/div/div[3]/div/div/div/div[1]/div[2]/h3")
          go_back_btn = driver.find_element(By.XPATH, "/html/body/div[7]/div/div/div/div/div[2]/div/div[1]/div/div[3]/div/div/div/div[1]/div[1]/a/i")
          go_back_btn.click()
          time.sleep(2)
        except NoSuchElementException as e:
          print("Not in the matter")
        finally:
          matterLbl = driver.find_element(By.XPATH, "/html/body/div[7]/div/div/div/div/div[2]/div/div[1]/div/div[3]/div/div/div/div[1]/div[2]/h2")
          print(matterLbl.text) # should always print "Matters"


```
